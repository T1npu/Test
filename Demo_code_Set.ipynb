```python
# 1. Khởi tạo Set và chứng minh tính duy nhất (Tự động lọc trùng)
my_set = {1, 2, 2, 3, 3, 3, "Python", "Python"}
print("Set sau khi khởi tạo:", my_set)

# 2. Chứng minh tính không thứ tự (Unordered) -> Không thể dùng Index
try:
    print(my_set[0])
except TypeError as e:
    print(f"Lỗi khi dùng Index: {e}")

# 3. Chứng minh Set có thể thay đổi (Mutable) thông qua add/remove
my_set.add("FPT")
my_set.remove(1)
print("Set sau khi thêm/xóa:", my_set)
```

```python
# So sánh hành vi lọc trùng giữa List và Set
sample_list = [1, 2, 2, 3]
sample_set = {1, 2, 2, 3}

print("List giữ nguyên phần tử trùng:", sample_list) 
print("Set tự động ép thành duy nhất:", sample_set)   

# So sánh cú pháp giữa Dictionary và Set (Cùng dùng dấu ngoặc nhọn)
sample_dict = {"key1": "value1", "key2": "value2"}  
print("Đây là Dict:", type(sample_dict))            
print("Đây là Set:", type(sample_set))               
```

```python
# Giả sử có 2 tập hợp kỹ năng của 2 ứng viên
candidate_A = {"Python", "SQL", "Excel"}
candidate_B = {"Java", "SQL", "C++"}

# 1. Phép GIAO (Intersection): Tìm kỹ năng cả hai đều có
print("1. Kỹ năng chung (&):", candidate_A.intersection(candidate_B))


# 2. Phép HỢP (Union): Gom toàn bộ kỹ năng không trùng lặp
print("2. Tổng hợp kỹ năng (|):", candidate_A.union(candidate_B))


# 3. Phép HIỆU (Difference): Kỹ năng chỉ ứng viên A có mà B không có
print("3. Kỹ năng riêng của A (-):", candidate_A.difference(candidate_B))

group_math = {"An", "Bình", "Cường"}
group_physics = {"Bình", "Duy", "An"}

# phép toán tập hợp tối ưu bằng mã C
both_classes = group_math & group_physics
print("4. Kỹ năng chung (&):",both_classes)
```

```python
import time

large_list = list(range(10000000))
large_set = set(range(10000000))
target = 9999999  # Phần tử nằm ở cuối danh sách

# Đo thời gian tìm kiếm trên LIST
start_time = time.time()
print(target in large_list)  # Kết quả: True
print(f"Thời gian List tìm kiếm: {time.time() - start_time:.6f} giây")


# Đo thời gian tìm kiếm trên SET
start_time = time.time()
print(target in large_set)   # Kết quả: True
print(f"Thời gian Set tìm kiếm: {time.time() - start_time:.6f} giây")
```

```python
import sys

# Tạo 1 List và 1 Set chứa cùng các số từ 0 đến 999
numbers_list = list(range(1000))
numbers_set = set(range(1000))

print(f"Kích thước List trong RAM: {sys.getsizeof(numbers_list)} bytes")
print(f"Kích thước Set trong RAM: {sys.getsizeof(numbers_set)} bytes")
```
